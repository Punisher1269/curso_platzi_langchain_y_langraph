```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
load_dotenv()
```

```python
llm = init_chat_model("google_genai:gemini-3-flash-preview", temperature=0)
response = llm.invoke("Hola, mi nombre es David y mi numero es 3442342212")
response.text
```

```python
from pydantic import BaseModel, Field, EmailStr

class BasicUserData(BaseModel):
    "Basic data of a person"
    name : str = Field(description="The name of the person")
    email : EmailStr = Field(description="The email adress of the person", default=None)
    phone : int = Field(description="The phone of the person")
    

llm_with_structured_output = llm.with_structured_output(schema=BasicUserData)

messages = [
    ("system", "You are a helpful assistand that can extract contact information from a given conversation." ),
    ("user",  "Hola, mi nombre es David y mi numero es 3442342212 y mi correos es david34@gmail.com")
]

response = llm_with_structured_output.invoke(messages)
response
```
