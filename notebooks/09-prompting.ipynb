```python
from langchain_core.prompts import PromptTemplate

template = """\
# Instrucciones para el asistente de la IA

Recuerda de hoy es {fecha_actual}

## Rol
Eres un experto en marketing digital con 10 años de experiencia

## Tarea
Analiza el siguiente texto de un anuncio y proporciona 3 sugerencias para mejorarlo, enfocate en el llamado a la accion

## Texto del anuncio
{texto_anuncio}

## Formato de salida
- *Sugerencia 1:* [Tu primera sugerencia]
- *Sugerencia 2:* [Tu segunda sugerencia]
- *Sugerencia 3:* [Tu tercera sugerencia]

"""

prompt_template = PromptTemplate.from_template(template)

prompt = prompt_template.format(fecha_actual='2026-05-14', texto_anuncio="Este es un anuncion de prueba" )

print(prompt)
```

```python
# Variables parciales o por defecto
import datetime
today = datetime.datetime.now().strftime("%Y-%m-%d")
prompt_template = PromptTemplate.from_template(template, partial_variables={"fecha_actual" : today})
prompt = prompt_template.format(texto_anuncio="Este es un anuncion de prueba" )
 
print(prompt)
```

```python
# Uso avanzado de prompts dinamicos con Jinja2
template="""\
    Your are a helpful that responds to the user's message

    {% if name %}
    The user's name is {name} and you can call them by that name
    {% endif -%}
    the user's message is: {{user_message}}
"""


prompt_template = PromptTemplate.from_template(template, template_format="jinja2")
prompt = prompt_template.format(user_message="Este es un mensaje de prueba" )
 
print(prompt)
```
