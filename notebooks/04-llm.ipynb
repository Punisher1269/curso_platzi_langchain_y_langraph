```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_anthropic import ChatAnthropic

anthropic = ChatAnthropic(model_name="claude-haiku-4-5-20251001", temperature=0)
anthropic.invoke("Hola, como estas ?")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

genai = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
response = genai.invoke("Hola, como estas ?")
response.pretty_print()
```

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain.chat_models import init_chat_model

msg_1 = SystemMessage(content="You are a helpful asistant that talks in Spanish")
msg_2 = HumanMessage(content="Ma llamo David")
msg_3 = AIMessage(content="Hola David, como estas ?")
msg_4 = HumanMessage(content="Como me llamo ?")

history = [msg_1, msg_1, msg_3, msg_4]

genai = init_chat_model("google_genai:gemini-3-flash-preview", temperature=0)
response = genai.invoke(history)
response.pretty_print()
```
