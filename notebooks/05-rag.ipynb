```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model


llm = init_chat_model("openai:gpt-4o", temperature=0)
```

```python
file_search_tool = {
    "type" : "file_search",
    "vector_store_ids" : ["vs_69fb5f76f630819199504fcfd2ad76bd"] # id de la base de datos vectorial
}
llm_with_tools = llm.bind_tools([file_search_tool])
```

```python
response = llm_with_tools.invoke("Como es el nombre del proyecto ?")

response.text
```

```python
response = llm_with_tools.invoke("Hola, como estas ?")

response.text
```
